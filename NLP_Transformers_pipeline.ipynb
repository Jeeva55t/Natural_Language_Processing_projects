```python
from transformers import pipeline

classifier = pipeline("text-classification")
result = classifier("I love using NLP models!")
print(result)
```

```python
ner = pipeline("ner", grouped_entities=True)
result = ner("Barack Obama was born in Hawaii.")
print(result)
```

```python
question_answerer = pipeline("question-answering")
question_answerer(
    question = "What is the full name of Ronaldo?",
    context ="Cristiano Ronaldo's full name is Cristiano Ronaldo dos Santos Aveiro. The CR7 nickname combines his initials (CR) with his jersey number (7).He is a Portuguese professional footballer known for his accomplishments and is often considered one of the greatest players of all time."
)
```

```python
summarizer = pipeline("summarization")
text = """
Cristiano Ronaldo's full name is Cristiano Ronaldo dos Santos Aveiro. The CR7 nickname combines his initials (CR) with his jersey number (7).He is a Portuguese professional footballer known for his accomplishments and is often considered one of the greatest players of all time.
"""
summarizer(text,max_length=50)
```

```python
from transformers import AutoTokenizer,AutoModelForSequenceClassification
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
sentence = "I love this product!"
inputs = tokenizer(sentence,return_tensors="pt")
```

```python
outputs = model(**inputs)
```

```python
probabilities = torch.softmax(outputs.logits,dim=1)
probabilities
```

```python
label_id = torch.argmax(probabilities).item()
label = model.config.id2label[label_id]
```

```python
print(f"Sentence :{sentence}")
print(f"Label :{label}")
print(f"Probabilities: {probabilities.tolist()}")
```
